```python
#1 극성 페이지 추출
def make_Emotion_dict():
    f = open(f"./call_rate_months.csv", 'r')
    file_lines = f.readlines()
    date = file_lines[0].split(",")
    rate = file_lines[1].split(",")
    date = date[:-4]
    rate = rate[:-4]
    date_dict = {date[idx]: rate[idx] for idx in range(len(rate))}
    target_date_dict = {}
    for i in date_dict:
        temp = i.replace("/",".")
        target_date_dict[temp] = date_dict[i]
    key_list = [key for key, value in target_date_dict.items()]
    value_list = [value for key, value in target_date_dict.items()]
    Emotion_dict = {"doveish":[], "hawkish":[]}
    for idx in range(len(key_list)-1):
        target_date = key_list[idx]
        target_rate = value_list[idx]
        compare_date = key_list[idx+1]
        compare_rate = value_list[idx+1]
        threshold_temp = float(target_rate) - float(compare_rate)
        threshold_temp = (threshold_temp**2)**0.5
        if threshold_temp > 0.001:
            if target_rate > compare_rate:
                Emotion_dict["doveish"] += [[target_date, compare_date]]
            else:
                Emotion_dict["hawkish"] += [[target_date, compare_date]]
    return Emotion_dict
```

```python
from ekonlpy import Mecab
import re
def normalization(word_list):
    word_li = []
    Train_list = []
    for w in word_list:
        temp = re.compile('[^ㄱ-ㅣ가-힣+]')
        result = temp.sub('', w[0])
        label = w[1]
        #print(result)
        #break
        mecab = Mecab()
        pos_tag = mecab.pos(result)
        temp = []
        for i in pos_tag:
            word = i[0]
            pos = i[1]
            if pos == "NNG" or pos == "VA" or pos == "VAX" or pos == "MAG" or pos == "VV":
                temp.append(word)
        temp_1 = ' '.join(temp)
        Train_list.append([temp_1,label])
    return Train_list
```

```python
import os

def select_MPB(target_date):
    start_date = "2013-05-09"
    def date_range(start, end):
        start = datetime.strptime(start, "%Y-%m-%d")
        end = datetime.strptime(end, "%Y-%m")
        before_one_day = end - relativedelta(days=30)
        dates = [(start + timedelta(days=i)).strftime("%Y-%m-%d") for i in range((before_one_day-start).days+1)]
        return dates

    dates = date_range(start_date, target_date)

    path = "./MPB_crawling_dataset"
    file_list = []
    if os.path.isdir(path):
        text_list = os.listdir(path)
        for i in text_list:
            if i[-3:] == "txt":
                target_file = i[:-4]
                if target_file in dates:
                    file_list.append(i)
    MPB_file = file_list[-1]                
    f = open(f"{path}/{MPB_file}", 'r')
    file_text = f.readline()
    return file_text

```

```python
import numpy as np

seed_word_dict = {"hawkish": ['높', '팽창', '인상', '매파', '성장', '투기_억제', '상승',  '인플레이션_압력', '증가', '위험_선호', '상회', '물가_상승', '과열', '금리_상승', '확장', '상방_압력', '긴축', '변동성_감소', '흑자', '채권_가격_하락', '견조', '요금_인상', '낙관', '부동산_가격_상승', '상향']
                  
          
                  ,"doveish":["낯", "축", "인하", "비둘기", "둔화", "악화", "하락", "회복_못하", "감소", "위험_회피", "하회", "물가_하락", "위축", "금리_하락", "침체", "하방_압력", "완화", "변동성_확대", "적자", "채권_가격_상승", "부진", "요금_인하", "비관", "부동산_가격_하락", "하향"]}

import numpy as np
from numpy import dot
from numpy.linalg import norm
from ekonlpy import Mecab
from gensim.models import Word2Vec
from gensim.models.phrases import Phrases, Phraser

# seed_model 구축
def build_Seed_Model(seed_word_dict):
    target = list(seed_word_dict.values())
    word_list = target[0] + target[1]
    phrases = Phrases([x.split() for x in word_list], min_count=1, threshold=1.1)
    ngram_word_list = [' '.join(phrases[x.split()]) for x in word_list]
    Seed_Model = Word2Vec([x.split() for x in ngram_word_list], vector_size=300, window=5, min_count=1)
    return Seed_Model

Seed_Model = build_Seed_Model(seed_word_dict)

# 넘파이 코사인 유사도 계산
def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))

def ngram2vec_train(Train_list):
    X_train = [i[0] for i in Train_list]

    # 2. 특성 추출 (형태소 분석 및 N-그램)
    mecab = Mecab()
    mecab_vocab = [mecab.pos(token) for token in X_train]
    vectors = []
    for vocab in mecab_vocab:
        for i in vocab:
            token = i[0]
            pos = i[1]
            if pos == "NNG" or pos == "VA" or pos == "VAX" or pos == "MAG" or pos == "VV":
                vectors.append(token)

    # 3. N-그램 추출
    phrases = Phrases([x.split() for x in vectors], min_count=1, threshold=1.1)
    ngram_X_train = [' '.join(phrases[x.split()]) for x in vectors]
    New_test_ngrams = New_test_ngrams[0]
    n_gram_train = []
    for idx in range(len(New_test_ngrams)):
        if len(New_test_ngrams)-idx == 4:
            target_0 = ngram_X_train[idx]
            target_1 = '_'.join(ngram_X_train[idx:idx+2])
            target_2 = '_'.join(ngram_X_train[idx:idx+3])
            target_3 = '_'.join(ngram_X_train[idx:idx+4])
            info = [target_0] + [target_1] + [target_2] + [target_3]
            n_gram_train += info
        elif len(New_test_ngrams)-idx == 3:
            target_0 = ngram_X_train[idx]
            target_1 = '_'.join(ngram_X_train[idx:idx+2])
            target_2 = '_'.join(ngram_X_train[idx:idx+3])
            info = [target_0] + [target_1] + [target_2]
            n_gram_train += info
        elif len(New_test_ngrams)-idx == 2:
            target_0 = ngram_X_train[idx]
            target_1 = '_'.join(ngram_X_train[idx:idx+2])
            info = [target_0] + [target_1]
            n_gram_train += info
        elif len(New_test_ngrams)-idx == 1:
            target_0 = ngram_X_train[idx]
            info = [target_0]
            n_gram_train += info
        else:
            target_0 = ngram_X_train[idx]
            target_1 = '_'.join(ngram_X_train[idx:idx+2])
            target_2 = '_'.join(ngram_X_train[idx:idx+3])
            target_3 = '_'.join(ngram_X_train[idx:idx+4])
            target_4 = '_'.join(ngram_X_train[idx:idx+5])
            info = [target_0] + [target_1] + [target_2] + [target_3] + [target_4]
            n_gram_train += info
        n_gram_train += info

    # 4. Ngram2Vec 모델 학습
    model = Word2Vec([x.split() for x in n_gram_train], vector_size=300, window=5, min_count=25)
    return model,phrases
model,phrases = ngram2vec_train(word_list)

# 극성 사전 구축
def Emotion(model,Seed_Model):
    New_Emotion_len_dict = {"hawkish":0, "doveish":0}
    New_Emotion_dict = {"hawkish":[], "doveish":[]}
    vector_list =  model.wv.index_to_key
    Seed_Model_list = Seed_Model.wv.index_to_key
    # hawkish seed_words, model.wv.index_to_key로 가져와서 평균값
    # dovish seed words 
    for vector in vector_list:
        hawkish_score = 0
        doveish_score = 0
        target_score = np.array(model.wv[vector])
        for seed_word in seed_word_dict["hawkish"]:
            hawkish_model_target = np.array(Seed_Model.wv[seed_word])
            hawkish_cosin = cos_sim(target_score, hawkish_model_target)
            hawkish_score += hawkish_cosin
        for seed_word in seed_word_dict["doveish"]:
            doveish_model_target = np.array(Seed_Model.wv[seed_word])
            doveish_cosin = cos_sim(target_score, doveish_model_target)
            doveish_score +=  doveish_cosin
        if hawkish_score//len(seed_word_dict["hawkish"]) > doveish_score//len(seed_word_dict["doveish"]):
            print(f"{vector} 해당 단어는 매파적입니다")
            New_Emotion_len_dict["hawkish"] += 1
            New_Emotion_dict["hawkish"] += [vector]
        elif hawkish_score == doveish_score:
            print(f"{vector} 해당 단어는 중립입니다 고려하지 않습니다")
        elif hawkish_score//len(seed_word_dict["hawkish"]) < doveish_score//len(seed_word_dict["doveish"]):
            print(f"{vector} 해당 단어는 비둘기적입니다")
            New_Emotion_len_dict["doveish"] += 1
            New_Emotion_dict["doveish"] += [vector]
    return New_Emotion_len_dict,New_Emotion_dict

New_Emotion_len_dict,New_Emotion_dict = Emotion(model,Seed_Model)


# 1. 데이터 준비
def model_result(New_Emotion_len_dict,New_Emotion_dict,phrases):
    # 새로운 입력 예측하기 위한 전처리 (형태소 분석 및 N-그램 변환)
    New_test_text = [file_text]
    New_test_ngrams = [(phrases[x.split()]) for x in New_test_text]

    hawkish_list = list(New_Emotion_dict["hawkish"])
    doveish_list = list(New_Emotion_dict["doveish"])

    hawkish_score = 0
    doveish_score = 0
    for i in New_test_ngrams[0]:    
        if i in hawkish_list:
            #print("매파적입니다")
            hawkish_score += 1
        elif i in doveish_list:
            #print("비둘기적입니다")
            doveish_score += 1
    if hawkish_score > doveish_score:
        return "매파적입니다"
    elif hawkish_score == doveish_score:
        return "중립적 입니다"
    elif hawkish_score < doveish_score:
        return "비둘기적입니다"
    
my_model_result = model_result(New_Emotion_len_dict,New_Emotion_dict,phrases)
print(my_model_result)
```
