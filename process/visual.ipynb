# Word Cloud

```python
# MPB
import os 
import re
from ekonlpy import Mecab 
import pandas as pd

file_list = os.listdir('./MPB_dataset/MPB_crawling_dataset/')  # 경로

MPB_pos = []


for file_name in file_list:

    with open('./MPB_dataset/MPB_crawling_dataset/' + file_name, 'r', encoding='utf-8') as f:  # 경로
        content = f.read()
        
        regex = re.compile('[^ㄱ-ㅣ가-힣+]')

        result = regex.sub('', content)

        mecab = Mecab()
        pos_tag = mecab.pos(result)
        
        for p in pos_tag:
            
            if (p[1] == "NNG" or p[1] == "VA" or p[1] == "VAX" or p[1] == "MAG" or p[1] == "VV"):
                MPB_pos.append(p[0])
        

print(MPB_pos)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
from PIL import Image
import numpy as np

c = Counter(MPB_pos)
```

```python
print(c)
```

```python
# MPB_워드 클라우드
wc = WordCloud(font_path='malgun', width=400, height=400, scale=2.0, 
               max_font_size=250, background_color="white")
gen = wc.generate_from_frequencies(c)
plt.axis("off")

plt.imshow(gen)
```

```python
# 워드 클라우드 파일로 추출
wc.to_file('법전_워드클라우드.png')
```

```python
# report
import os 
import re
from ekonlpy import Mecab 
import pandas as pd

file_list = os.listdir('./dataset/report/dataset/')  # 경로

report = []

#print(file_list)

for file_name in file_list:
    files = os.listdir('./dataset/report/dataset/' + file_name)
    #print(files)

    for file in files:
        #print(file)
        with open('./dataset/report/dataset/' + file_name + '/' + file, 'r', encoding='utf-8') as f:  # 경로
            content = f.read()
            table = content.maketrans('제목요약그림내용자료', '          ')
            new_con = content.translate(table)
            report.append(new_con)
```

```python
report[0]
```

```python
report_pos = []

for r in report:
    regex = re.compile('[^ㄱ-ㅣ가-힣+]')

    result = regex.sub('', r)

    mecab = Mecab()
    pos_tag = mecab.pos(result)
    
    for p in pos_tag:
        
        if (p[1] == "NNG" or p[1] == "VA" or p[1] == "VAX" or p[1] == "MAG" or p[1] == "VV"):
            report_pos.append(p[0])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
from PIL import Image
import numpy as np

c = Counter(report_pos)
```

```python
print(c)
```

```python
# report_워드 클라우드
wc = WordCloud(font_path='malgun', width=400, height=400, scale=2.0, 
               max_font_size=250, background_color="white")
gen = wc.generate_from_frequencies(c)
plt.axis("off")

plt.imshow(gen)
```

```python
# MPB
import os 
import re
from ekonlpy import Mecab 
import pandas as pd

file_list = os.listdir('./MPB_dataset/MPB_crawling_dataset/')  # 경로

date = []
sentence = []

for file_name in file_list:
    with open('./MPB_dataset/MPB_crawling_dataset/' + file_name, 'r', encoding='utf-8') as f:  # 경로
        
        date.append(file_name[:-4])
        sentence.append(f.read())

    #break
#print(date)
#print(sentence)

df_1 = pd.DataFrame(date, columns=['date'])
df_2 = pd.DataFrame(sentence, columns=['sentence'])

df = pd.concat([df_1, df_2], axis=1)

df
```

```python
def count_sentences(text):
    return text.count('.')
```

```python
df['sentence_count'] = df['sentence'].apply(count_sentences)
```

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
```

```python
import matplotlib.pyplot as plt

grouped = df.groupby('year')['sentence_count'].sum()

plt.figure(figsize=(10,5))
grouped.plot()

plt.ylabel('Number of Sentences')
plt.title('Number of Sentences per Year')
plt.show()
```

```python
# edaily

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import dates
import numpy as np

edaily = pd.read_csv('./dataset/edaily.csv')
edaily['Date'] = pd.to_datetime(edaily['Date'])
monthly = edaily.groupby(edaily['Date'].dt.to_period('M')).count()
monthly.drop(['Date', 'Content'], axis=1, inplace=True)
monthly = monthly.rename(columns={'Title':'The Number Of News'})
monthly.plot()
plt.xlabel('Year')
plt.ylabel('The Number Of News')
plt.show()
```
